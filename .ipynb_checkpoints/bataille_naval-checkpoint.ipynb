<font color="RED" size="+1">**[Réponse 1]**</font> **Le nombre de configuration possible :**

la born sup pour le nombre de configuration possible est de 77 milliards.
#### Raisonnement : 
  on considere une grille vide a chaque placement d'un bateau, ce qui fait : 

  si par exemple on veut placer un porte-avions( 5 cases ). Les lignes et les colonnes contiennent 10 cases, ce qui veut     dire qu'on peut le placer dans 6 cases différentes. Ce qui donne 6 * 10 pour les lignes de la grille ainsi que pour les     colonnes.
  Donc ( 6*10 ) *2 = 120 configurations possibles pour placer un porte avions.
  on raisonne pareillement pour les autres types de bateau, ce qui donne :

  - porte-avions(5 cases): ( 6 * 10 ) *2 = 120

  - croiseur(4 cases): ( 7 * 10 ) * 2 = 140

  - contre-torpilleurs(3 cases): ( 8 * 10 ) *2 = 160

  - sous-marin(3 cases): ( 8 * 10 ) *2 = 160

  - torpilleur(2 cases): ( 9 * 10 ) *2 = 180 

120 *140 *160 *160 * 180 =   7,74144e10

##### remarque:
  cette borne ne sera jamais atteinte car on considère a chaque fois une grille vide. Mais ca reste une borne supérieure     qui ne sera jamais dépasser.

<font color="RED" size="+1">**[Réponse 1]**</font> **Poser les hypothése et calculer l'esperance :**

#### l'hpothése :
    l'hypothése serais de dire que tout les coups son equipobable et qu'ils sont independent
#### calcule de l'espérance :
    tb : la somme des taille du bateau 
    p : probabilité de reussir un coup
    q : probabilité de rater le coup 
    
    On note Pn la probabilité que le joueur gange a la n eme partie( apres n essaie )
    donc pour que le joueur gange une partie il faut qu'il gange tb-1 fois a la n-1 eme partie et qu'il gange la n eme  
    partie la probablité d'une partie serait  p**tb * q**(n-tb)
    donc comme il y'a C(n-1,tb-1) facon de ganger tb-1 partie sur n-1 essye partie on a :
                
                    Pn = C(n-1,tb-1) * p**tb * q**(n-tb)
